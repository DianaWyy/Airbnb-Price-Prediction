```python
import numpy as np
import pandas as pd
import sklearn
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("../train1.csv")
df_test = pd.read_csv("../test1.csv")
```

```python
from sklearn.linear_model import LinearRegression
## Linear Regression
X_train = df[['price', 'pca_component 0', 'pca_component 1', 'pca_component 2', 'pca_component 3', 'pca_component 4', "pca_component 5", "pca_component 6", "pca_component 7", "pca_component 8", "pca_component 9", "pca_component 10"]]
y_train = df['price']

# Fitting linear regression
reg = LinearRegression().fit(X_train, y_train)

# R squared value
print('reg score: ', reg.score(X_train, y_train))

# Coefficients
print('reg coef: ', reg.coef_)

# Intercept
print('reg intercept: ', reg.intercept_)


Coeff = pd.DataFrame(columns=["Variable","Coefficient"])
Coeff["Variable"]=X_train.columns
Coeff["Coefficient"]=reg.coef_
Coeff.sort_values("Coefficient")
```

```python
# Predicting 
X_test = df_test[['price', 'pca_component 0', 'pca_component 1', 'pca_component 2', 'pca_component 3', 'pca_component 4', "pca_component 5", "pca_component 6", "pca_component 7", "pca_component 8", "pca_component 9", "pca_component 10"]]
y_test = df_test['price']
y_pred = reg.predict(X_test)

# Calculate RMSE
from sklearn import metrics
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
rmse
```

```python
# Taking a closer look at the estimates
import statsmodels.api as sm

X2 = sm.add_constant(X_train)
est = sm.OLS(y_train, X2)
est2 = est.fit()
print(est2.summary())
```

```python
from sklearn.preprocessing import PolynomialFeatures

# Polynomial Regression
poly = PolynomialFeatures(degree = 2) 
X_poly = poly.fit_transform(X_train) 
  
poly.fit(X_poly, y_train) 
lin = LinearRegression() 
lin.fit(X_poly, y_train) 

# R squared value
print('polyreg score: ', lin.score(X_poly, y_train))

# Coefficients
print('polyreg coef: ', lin.coef_)

# Intercept
print('polyreg intercept: ', lin.intercept_)
```

```python
# Calculate RMSE
poly_y_pred = lin.predict(poly.fit_transform(X_train))
poly_rmse = np.sqrt(metrics.mean_squared_error(y_train, poly_y_pred))
poly_rmse
```

```python
# Taking a closer look at the estimates
X3 = sm.add_constant(X_poly)
est3 = sm.OLS(y_train, X3)
est4 = est3.fit()
print(est4.summary())
```
