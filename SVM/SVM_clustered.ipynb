```python
import numpy as np
import pandas as pd
import sklearn
from matplotlib import pyplot as plt
```

```python
from sklearn.svm import LinearSVC, SVR, SVC
from sklearn.neighbors import KNeighborsClassifier
# For Storing Models
import pickle
# For timing
import time
from datetime import timedelta
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_squared_log_error, r2_score
```

## For clustered data

```python
df = pd.read_csv("../Unsupervised_learning/train1_clustered.csv", header=0, index_col=0)
df_test = pd.read_csv("../test1.csv")
```

```python
df.head()
```

```python
df_test = df_test.drop(['Unnamed: 0'], axis=1)
```

```python
df_test.head()
```

```python
x_train = df.drop(['price', 'cluster', 'price_category'], axis=1)
y_train = df['cluster']
```

```python
x_test = df_test.drop(['price'], axis=1)
```

```python
knn = KNeighborsClassifier().fit(x_train, y_train)
y_test = knn.predict(x_test)
df_test['cluster'] = y_test
```

```python
#Train the SVRs with different kernels
clf = [SVR(kernel='linear'),
       SVR(kernel='rbf'),
       SVR(kernel='poly')]
```

```python
df_test.head()
```

```python
df.head()
```

```python
for k in range(3): # for each kernal
    original_price = []
    predicted_price = []
    for i in range(3): # for each cluster
        train = df[df['cluster'] == i]
        test = df_test[df_test['cluster'] == i]
        train_x = train.drop(['cluster', 'price', 'price_category'],axis=1)
        train_y = train['price']
        clf[k].fit(train_x, train_y)
        original_price.append(test['price']) #
        test_x = test.drop(['cluster', 'price'], axis=1)
        test_y = clf[k].predict(test_x)
        predicted_price.append(test_y) #
    original = np.concatenate(original_price)
    predicted = np.concatenate(predicted_price)
    rmse = np.sqrt(mean_squared_error(predicted, original))
    mae = mean_absolute_error(predicted, original)
    log_mse = mean_squared_log_error(predicted, original)
    r2 = r2_score(predicted, original)
    print("rmse is: ", rmse)
    print("mae is: ", mae)
    print("log_mse is: ", log_mse)
    print("R^2 value using score fn: %.3f" % r2)
```
